```python
for games in open('files/games.txt'):
    print(games, end='')
    # end will remove extra \n
```

```python
cities, years = [], []
for games in open('files/games.txt'):
    vals = games.split()
    # combine all the values except last one
    city = ' '.join(vals[:-1])
    # last val is year, this -1 will go to last spot
    # remove parantesis around year as well
    year = vals[-1].strip('()')
    # end will remove extra \n
    print("City: ", city, " year: ", year)
    cities.append(city)
    years.append(year)
```

```python
# create a geolocation service
geo = geopy.geocoders.Nominatim(user_agent='dmitriy-efficient-prog')
```

```python
locations = {}
for city in cities:
    locations[city] = geo.geocode(city.split('/')[0])
    print("City: ", city, " Location: ", locations[city])
    print("City: ", city, " Lat: ", locations[city].latitude, " Lon: ", locations[city].longitude)
```

```python
pp.figure(figsize=(10,5))

axes = pp.axes(projection = cartopy.crs.PlateCarree())
axes.add_feature(cartopy.feature.COASTLINE, linewidth=0.5)
axes.add_feature(cartopy.feature.BORDERS, linewidth=0.5)

for pos in locations.values():
    axes.plot(pos.longitude, pos.latitude, 'b*b')
```

```python
print("=== sorted ===")
for city in sorted(cities[:5]):
    print(city)

print("=== reversed and sorted ===")
for city in reversed(sorted(cities[:5])):
    print(city)
    
print("=== zipped and sorted ===")
for city, year in sorted(zip(cities[:5], years[:5])):
    print(city, year)    
```
